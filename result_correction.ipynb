```python
import pandas as pd
```

```python
result = pd.read_csv('result_2418.csv')
```

```python
result_500 = pd.read_csv('result_500.csv')
```

```python
forecast_2018 = result[(result['year']==2018)&(result['data_type']=='biomass_forecast')]
```

```python
biomass_demand_2018 = result[(result['year']==2018)&(result['data_type']=='biomass_demand_supply')]
pellet_demand_2018 = result[(result['year']==2018)&(result['data_type']=='pellet_demand_supply')]
```

```python
forecast_2018['value'] = 1.2*forecast_2018['value']
forecast_2018['year'] = 2019
pellet_demand_2018['value'] = 1.2*pellet_demand_2018['value'] 
biomass_demand_2018['value'] = 1.2*biomass_demand_2018['value']
pellet_demand_2018['year'] = 2019
biomass_demand_2018['year'] = 2019
```

```python
concated_df = pd.concat([result,forecast_2018,pellet_demand_2018,biomass_demand_2018])
```

```python
concated_df.groupby(['year','data_type']).agg(set)
```

```python
concated_df[concated_df['value']>=0.001]
```

```python
remove_index = concated_df[((concated_df['data_type']=='biomass_demand_supply')|(concated_df['data_type']=='pellet_demand_supply'))&(concated_df['value']<=0.001)].index
```

```python
 concated_df.drop(index=remove_index).to_csv('sumbimmision.csv',index=False)
```
