```python
import numpy as np
import pandas as pd
```

```python
distance_df = pd.read_csv("../waste_management/dataset/Distance_Matrix.csv")
distance_np = distance_df.values
distance_np
```

```python
# Define the distance matrix (distance from each demand point to each potential facility)
distance_matrix = distance_np

# Define the fixed costs of opening facilities
facility_costs = np.array([0 for i in range(2418)])

def greedy_facility_location(distance_matrix, facility_costs):
    num_facilities = len(facility_costs)
    num_demand_points = distance_matrix.shape[0]
    
    # Initialize an empty list to store the chosen facility for each demand point
    assignment = [-1] * num_demand_points
    
    for demand_point in range(num_demand_points):
        min_cost = float('inf')
        chosen_facility = -1
        
        for facility in range(num_facilities):
            cost = distance_matrix[demand_point][facility] + facility_costs[facility]
            if cost < min_cost:
                min_cost = cost
                chosen_facility = facility
        print(cost) 
        assignment[demand_point] = chosen_facility
    
    return assignment

# Find the facility assignments using the Greedy Algorithm
facility_assignment = greedy_facility_location(distance_matrix, facility_costs)
print("Facility assignments:", facility_assignment)
```
