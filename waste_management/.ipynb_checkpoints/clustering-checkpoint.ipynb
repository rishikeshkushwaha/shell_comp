```python
import plotly.express as px
import pandas as pd

df = pd.read_csv("dataset/Biomass_History.csv")

fig = px.scatter_geo(df,lat='Latitude',lon='Longitude', hover_name="Index",color="Index")
fig.update_layout(title = 'World map', title_x=0.5)
fig.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np

# Sample data
X = np.array(df[['Latitude','Longitude']].values)
weights = np.array(df['2010'])  # Example weights

# Custom distance metric
def weighted_euclidean(u, v, weights):
    return np.sqrt(np.sum(weights * (u - v)**2))

# Custom k-means with weighted distance metric
class WeightedKMeans(KMeans):
    def __init__(self, n_clusters=8, weights=None, **kwargs):
        self.weights = weights
        super().__init__(n_clusters=n_clusters, **kwargs)

    def _distance(self, x, y):
        return weighted_euclidean(x, y, self.weights)

# Perform weighted k-means clustering
n_clusters = 25
weighted_kmeans = WeightedKMeans(n_clusters=n_clusters, weights=weights)
labels = weighted_kmeans.fit_predict(X)

print("Cluster labels:", labels)
```

```python
labels
```

```python

fig = px.scatter_geo(df,lat='Latitude',lon='Longitude', hover_name="Index",color=labels)
fig.update_layout(title = 'World map', title_x=0.5)
fig.show()
```

```python
cluster_centers = weighted_kmeans.cluster_centers_
```

```python
cluster_centers
```

```python
import plotly.graph_objects as go

lat_x = [x[0] for x in cluster_centers]
long_y = [x[1] for x in cluster_centers]
fig1 = px.scatter_geo(df,lat='Latitude',lon='Longitude', hover_name="Index",color=labels)

fig2 = px.scatter_geo(lat=lat_x,lon=long_y,)
# fig.update_layout(title = 'World map', title_x=0.5)
fig = go.Figure(data = fig1.data + fig2.data)
fig.show()
```

```python
delta = 0.08061
factor =1
lat_upper = np.array(lat_x)+factor*delta
lat_lower = np.array(lat_x)-factor*delta
long_right = np.array(long_y)+factor*delta
long_left = np.array(long_y)-factor*delta
```

```python
lat_upper-lat_lower,(long_right-long_left)/delta
```

```python
def potential_depot(x):
    for lower, upper in zip(lat_lower,lat_upper):
            if lower <= x['Latitude'] <= upper:
                for left, right in zip(long_left,long_right):
                        if left <= x['Longitude'] <= right:
                            return True
    return False
```
